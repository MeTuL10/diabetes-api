```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

from keras.models import Sequential,load_model
from keras.layers import Dense, Dropout
```

```python
predictor=load_model("models\\diabetes_predictor")
```

```python
df = pd.read_csv('dataset\\diabetes_prediction_dataset.csv')
```

```python
f=open("encodedlabels.txt","r")
txt=f.readline()
le1=LabelEncoder()
c1=txt.split(", ")
c1[-1]=c1[-1][:-1]
print(c1)

txt=f.readline()
le2=LabelEncoder()
c2=txt.split(", ")
c2[-1]=c2[-1]
print(c2)

le1.classes_=c1
le2.classes_=c2
```

```python
df.head()
```

```python
df["gender"]=le1.transform(df["gender"])
df["smoking_history"]=le2.transform(df["smoking_history"])
```

```python

test=df[df["diabetes"]==1]
test=test.iloc[:,:-1]
test
```

```python
a=predictor.predict(test)
```

```python
a.min(),a.max()
print(len(a))
print(len([i for i in a if i>0.1]))
```
